### Traženje elementa u listi

```c++
#include <iostream>
#include <stdio.h>
#include <stdlib.h>
#include <string.h>
//#include <math.h>
using namespace std;
```

```c++
typedef struct c{
    int broj;
    struct c *sljed;
}cvor;
```

```c++
cvor *novicvor(int broj) {
    cvor *pom = (cvor *)malloc(sizeof(cvor));
    pom->broj = broj;
    pom->sljed = 0;
    return pom;
}
```

```c++
int dodaj(cvor *glava, int broj) {
    while (glava->sljed !=0)
        glava = glava->sljed;
    glava->sljed = novicvor(broj);
    return 1;
}
```

```c++
void print(cvor *glava) {
    while (glava != NULL) {
        cout << glava->broj << " ";
        glava = glava->sljed;
    }
}
```

```c++
int niz[] = {4, 8, 12, 32, 34};

cvor *glava = NULL;

int size = sizeof(niz)/sizeof(int);

glava = novicvor(niz[0]);
for(int i=1; i<size; i++){
    dodaj(glava, niz[i]);
}

print(glava);
```
